```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
data = pd.read_csv('TimeSeries/international-airline-passengers.csv')
```

```python
data = data.drop(['Month'], axis=1)
data.columns = [0]
```

```python
data
```

```python
data_array = data.as_matrix()[:-1]
%xdel data
```

```python
plt.plot(data_array)
plt.show()
print(data_array.shape)
```

```python
np.random.seed(7)
```

```python
orig = data_array.copy()
scaler = MinMaxScaler(feature_range=(0, 1))
data_array = scaler.fit_transform(data_array)
```

```python
plt.plot(data_array)
plt.show()
print(data_array.shape)
```

```python
#80% del dataset = ceil( 144*0.8 ) = 116 + 1
#10% CV = 14 + 1
#10% test = 14
```

```python
train_array = data_array[0:117]
cv_array = data_array[116:131]
test_array = data_array[131:144]

print(train_array.shape)
print(cv_array.shape)
print(test_array.shape)
print('---------------------')
print(train_array[116], cv_array[0])
print(cv_array[13], test_array[0])
```

```python
def create_dataset(data):
    dataX, dataY = [], []
    for i in range(len(data)-1):
        a = data[i:(i+1), 0]
        dataX.append(a)
        dataY.append(data[i + 1, 0])
    return np.array(dataX), np.array(dataY)
```

```python
X_train, Y_train = create_dataset(train_array)
X_cross_val, Y_cross_val = create_dataset(cv_array)
X_test, Y_test = create_dataset(test_array)

print(X_train.shape, Y_train.shape)
print(X_cross_val.shape, Y_cross_val.shape)
print(X_test.shape, Y_test.shape)
```

```python
np.save('TimeSeries/orig', orig)

np.save('TimeSeries/X_train', X_train)
np.save('TimeSeries/Y_train', Y_train)

np.save('TimeSeries/X_cross_val', X_cross_val)
np.save('TimeSeries/Y_cross_val', Y_cross_val)

np.save('TimeSeries/X_test', Y_test)
np.save('TimeSeries/Y_test', Y_test)
```
