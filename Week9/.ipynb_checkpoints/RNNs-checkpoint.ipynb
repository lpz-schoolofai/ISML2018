# Leemos y Preprocesamos datos

```python
with open(path+'names', 'r') as f:
    file = f.readlines()
    data = [" "+name[:-1] for name in file] #AGREGAMOS UN ESPACIO
```

```python
print(len(data))
data
```

```python
###COMPARACIÓN LIST COMPREHENSION###
tokens =  list(set([letter for name in data for letter in name]))

tokens1 = []
for name in data:
    for letter in name:
        tokens1.append(letter)
tokens1 = set(tokens1)

print(len(tokens), len(tokens1))
assert len(tokens) == len(tokens1)
```

```python
token_to_id = dict( [(token, i) for (i, token) in enumerate(tokens)] )

print(token_to_id)
print()
print(sorted(list(token_to_id.keys())))
```

```python
###APLICAMOS GET A CADA CARACTER DE LA CADENA, PARA SACAR EL ÍNDICE##
print(len(data[0]))
list(map(token_to_id.get, data[0]))
```

```python
###CREAMOS LA MATRIZ X###
max_len = max(map(len, data))
print(max_len)
X_temp = np.zeros((len(data), max_len), dtype=np.int32) ##Tamaño fijo de cadenas

for i, name in enumerate(data):
    if i < 5: print(name)
    X_temp[i, :len(name)] = list(map(token_to_id.get, name))
X = X_temp.T.copy()

print(X.shape)
print(X[:, :5].T)
```

# El Modelo

```python
import keras
from keras.layers import Concatenate,Dense,Embedding
import tensorflow as tf
import matplotlib.pyplot as plt

rnn_num_units = 64
embedding_size = 16
n_tokens = len(tokens)

#Embede el caracter representado por el diccionario de 55 valores 
#a un espacio de 16 dimensiones
embed_x = Embedding(n_tokens, embedding_size)

#[x_t,h_t-1]->h_t
get_h_next = Dense(rnn_num_units, activation='elu') #W, U

#[h_t]-> y_hat
get_probas = Dense(n_tokens, activation='softmax') #V
```

```python
def rnn_one_step(x_t, h_t):
    """
    x_t = (m,) = (7944,)
    h_t = (m, rnn_num_units) = (7944, 64)
    """
    #convertir id único de caracter a embedding
    #x_T: (7944,) -> reshape([-1,1]) = (1, 7944) -> [:,0] = (7944,64)
    x_t_emb = embed_x( tf.reshape(x_t, [-1,1]) )[:,0] #(7944,16)
    #1er contexto
    
    #Concatenar x embedido y h_{t-1} -> horizonaltmente
    x_and_h = tf.concat([x_t_emb, h_t], 1) #(7944, 65)
    
    #computar h_t
    h_next = get_h_next(x_and_h) #(7944, 64)
    
    #computar y_hat_t
    output_probas = get_probas(h_next) #(7944, 55)
    
    return output_probas,h_next
```

```python
MAX_LENGTH = 16 #Maxima longitud de la cadena

input_sequence = tf.placeholder('int32',(MAX_LENGTH,None)) #(16,7944)
batch_size = tf.shape(input_sequence)[1]

predicted_probas = []
h_prev = tf.zeros([batch_size,rnn_num_units]) #inicializamos en cero

for t in range(MAX_LENGTH):
    #La primera letra de cada nombre del batch
    x_t = input_sequence[t] #(m,)
    probas_next,h_next = rnn_one_step(x_t,h_prev)
    
    h_prev = h_next
    predicted_probas.append(probas_next)
    
predicted_probas = tf.stack(predicted_probas) #(16, 7944, 55)
```

```python
###LOSS & GRADS###
#(16, 7944, 55)
predictions_matrix = tf.reshape(predicted_probas[:-1],[-1,len(tokens)])
# -> (15*7944, 55) eliminamos el ultimo caracter de la pred

#(16*7944)
answers_matrix = tf.one_hot(tf.reshape(input_sequence[1:],[-1]), n_tokens) #One-hot-encoding
# -> (15*7944, 55) eliminamos el primer caracter y volvemos one hot
```

```python
loss = tf.reduce_mean(tf.reduce_sum(-answers_matrix*tf.log(tf.clip_by_value(predictions_matrix,1e-10,1.0)), reduction_indices=[1]))

optimize = tf.train.AdamOptimizer().minimize(loss)
```

# Entrenamiento de la Red

```python
from IPython.display import clear_output
from random import sample
s = keras.backend.get_session()
s.run(tf.global_variables_initializer())
history = []
```

```python
for i in range(40):
    loss_batch = []
    for batch in range(int(X.shape[1]/32)+1):        
        batch_x = X[:, batch*32:(batch+1)*32]
        
        loss_b,_ = s.run([loss,optimize],{input_sequence:batch_x})
        loss_batch.append(loss_b)


    history.append(sum(loss_batch)/len(loss_batch))
    if (i+1)%10==0:
        clear_output(True)
        plt.plot(history,label='loss')
        plt.legend()
        plt.show()

assert np.mean(history[:10]) > np.mean(history[-10:]), "RNN didn't converge."
```

```python
x_t = tf.placeholder('int32',(None,))
h_t = tf.Variable(np.zeros([1,rnn_num_units],'float32'))

next_probs,next_h = rnn_one_step(x_t,h_t)
```

```python
def generar(semilla=' ',max_length=MAX_LENGTH):

    x_sequence = [token_to_id[token] for token in semilla]
    s.run(tf.assign(h_t, h_t.initial_value))
    
    #Iniciar con la semilla si existe
    for ix in x_sequence[:-1]:
         s.run(tf.assign(h_t, next_h),{x_t:[ix]}) #Recorremos la cadena mandando x_t
    
    #start generating
    for _ in range(max_length-len(semilla)):
        x_probs,_ = s.run([next_probs,tf.assign(h_t,next_h)],{x_t:[x_sequence[-1]]}) #Usamos y_hat_t-1 como x_t
        x_sequence.append(np.random.choice(n_tokens,p=x_probs[0]))
        
    return ''.join([tokens[ix] for ix in x_sequence])
```

```python
for _ in range(10):
    print(generar().strip())
```
